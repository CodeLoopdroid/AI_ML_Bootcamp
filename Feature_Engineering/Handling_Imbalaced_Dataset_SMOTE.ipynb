```python
"""
SMOTE (Synthetic Minority Over-sampling Technique)

Unlike simple upsampling, which just duplicates existing minority class samples, SMOTE generates new synthetic points. It does this by interpolating between existing minority samples that are close to each other in feature space. The new points are slightly varied but preserve the overall distribution of the data, helping the model learn better without simply copying the same points.
"""

from sklearn.datasets import make_classification
```

```python
x , y = make_classification(n_samples=1000, n_features= 2 , weights=[0.9], random_state=12,  n_clusters_per_class= 1, n_redundant=0)
```

```python
import pandas as pd

x = pd.DataFrame(x, columns = ['fea_1' , 'fea_2'])
y = pd.DataFrame(y , columns = ['target'])

df = pd.concat([x , y], axis = 1 ).reset_index(drop =True)
```

```python
from imblearn.over_sampling import SMOTE # Synthetic Minority Over-Sampling Technique.
```

```python
X , y = SMOTE(random_state=42).fit_resample(df[['fea_1', 'fea_2']], df['target'])
```

```python
x = pd.DataFrame(X, columns = ['fea_1' , 'fea_2'])
y = pd.DataFrame(y , columns = ['target'])

df = pd.concat([x , y], axis = 1 ).reset_index(drop =True)
```
